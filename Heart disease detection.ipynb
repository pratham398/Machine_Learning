## 1st Split(int x and y)

Splitting Data into Features and Target

```python
## fhs_cont contains only float values
fhs_cont.info()
```

separate array into input (x) and output(y) components

```python
x=fhs_cont
x.shape
```

creating the response/outcome matrix

```python
y=df_fhs['TenYearCHD']
y.shape
```

##**The Synthetic Minority Oversampling(SMOTE)**

**Applies SMOTE to balance the classes in the dataset by oversampling the minority class.**

```python
## Load the SMOTE module
from imblearn.over_sampling import SMOTE
```

```python
## creating test and train and SMOTE the X matrices
os=SMOTE(random_state=0)
x_os,y_os= os.fit_resample(x,y)
```

```python
## dimension of the matrices
x_os.shape,y_os.shape
```

```python
## distribution of categories of target variable
y_os.value_counts(normalize=True)
```

```python
## visua;ize the distribution-seaborn
sns.countplot(x=y_os)
plt.show()
```

```python
df_os=pd.DataFrame(x_os)
df_os['TenYearCHD']=y_os
```

```python
plt.figure(figsize=(10,7))
sns.scatterplot(data=df_os,x='BMI',y='cigsPerDay',hue='TenYearCHD')
plt.show()
```

##2nd split(train & test)

```python
## Import the 'train_test_split'function
from sklearn.model_selection import train_test_split
```

```python
## split data into train and test
x_train,x_test,y_train,y_test=train_test_split(x_os,y_os,test_size=0.3,random_state=0)
```

## Scaling the values

```python
## Importing 'standardscaler'function
from sklearn.preprocessing import StandardScaler
```

```python
## scaling train features for optimal model performance
## meaning there is no big data values difference
sc_train=StandardScaler().fit(x_train)
x_train_sc=sc_train.transform(x_train)
```

## training the model

```python
## create an instance of the Logistic Regression Model
from sklearn.linear_model import LogisticRegression
# #liblinear is particularly effective for small datasets and is the default
# choice for logistic regression.
model_lr=LogisticRegression(solver='liblinear')
```

```python
## use train data to  train the model features
model_lr.fit(x_train_sc,y_train)
```

## testing the model

```python
## scaling test featurees for optimal model performance
sc_test=StandardScaler().fit(x_test)
x_test_sc=sc_test.transform(x_test)
```

```python
## test the performance of the model on the test data
result=model_lr.score(x_test_sc,y_test)
result*100
```

```python
## test the performance of the model on the test data
result=model_lr.score(x_train_sc,y_train)
result*100
```
